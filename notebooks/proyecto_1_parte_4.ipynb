```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
df = pd.read_csv('../data/retail_sales_dataset.csv')

# Convertir la columna 'Date' a tipo datetime para su análisis
df['Date'] = pd.to_datetime(df['Date'])

# 1. Cálculo de estadísticas descriptivas
print("Estadísticas descriptivas:")
print(df[['Quantity', 'Price per Unit', 'Total Amount', 'Age']].describe())

# 2. Histogramas y Boxplots
# Configurar la figura para que contenga múltiples subgráficos
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Histogramas
sns.histplot(df['Quantity'], kde=False, color='blue', ax=axes[0, 0])
axes[0, 0].set_title('Distribución de la cantidad')

sns.histplot(df['Price per Unit'], kde=False, color='green', ax=axes[0, 1])
axes[0, 1].set_title('Distribución del precio por unidad')

# Boxplots
sns.boxplot(x=df['Quantity'], ax=axes[1, 0], color='blue')
axes[1, 0].set_title('Boxplot de Cantidad')

sns.boxplot(x=df['Price per Unit'], ax=axes[1, 1], color='green')
axes[1, 1].set_title('Boxplot de precio por unidad')

plt.tight_layout()
plt.show()

# 3. Gráfico de líneas para las tendencias de ventas (Total Amount)
df_grouped = df.groupby('Date')['Total Amount'].sum().reset_index()

plt.figure(figsize=(10,6))
plt.plot(df_grouped['Date'], df_grouped['Total Amount'], color='purple')
plt.title('Tendencias de Ventas a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Total de Ventas')
plt.xticks(rotation=45)
plt.show()

# 4. Gráficos de dispersión
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Dispersión entre 'Quantity' y 'Total Amount'
sns.scatterplot(x=df['Quantity'], y=df['Total Amount'], ax=ax[0], color='red')
ax[0].set_title('Cantidad vs Total de Venta')

# Dispersión entre 'Age' y 'Total Amount'
sns.scatterplot(x=df['Age'], y=df['Total Amount'], ax=ax[1], color='orange')
ax[1].set_title('Edad del Cliente vs Total de Venta')

plt.tight_layout()
plt.show()

# 5. Combinación de histogramas y boxplots para las ventas (Total Amount)
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# Histograma de 'Total Amount'
sns.histplot(df['Total Amount'], kde=False, color='teal', ax=axes[0])
axes[0].set_title('Distribución de Total de Venta')

# Boxplot de 'Total Amount'
sns.boxplot(x=df['Total Amount'], ax=axes[1], color='teal')
axes[1].set_title('Boxplot de Total de Venta')

plt.tight_layout()
plt.show()
```
