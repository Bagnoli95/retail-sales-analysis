```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset desde un archivo CSV
data_url = "../data/2019.csv"  # Reemplaza con la ruta real o URL del archivo CSV
df = pd.read_csv(data_url)

# Mostrar las primeras filas del dataframe
print(df.head())

# Explorar las columnas disponibles y el tipo de datos
print(df.info())
```

```python
# Lista de columnas para crear histogramas
variables = ['Score', 'GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices']

# Crear histogramas
for variable in variables:
    plt.figure(figsize=(8, 5))
    plt.hist(df[variable], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title(f'Histograma de {variable}')
    plt.xlabel(variable)
    plt.ylabel('Frecuencia')
    plt.show()
```

```python
# Crear boxplots
for variable in variables:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df[variable], color='lightgreen')
    plt.title(f'Boxplot de {variable}')
    plt.show()
```

```python
# Crear gráficos combinados (histograma y boxplot) para cada variable
for variable in variables:
    fig, axs = plt.subplots(2, 1, figsize=(8, 8))

    # Histograma
    axs[0].hist(df[variable], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
    axs[0].set_title(f'Histograma de {variable}')
    axs[0].set_xlabel(variable)
    axs[0].set_ylabel('Frecuencia')

    # Boxplot
    sns.boxplot(data=df[variable], ax=axs[1], color='lightgreen')
    axs[1].set_title(f'Boxplot de {variable}')

    plt.tight_layout()
    plt.show()
```

## ANALISIS
### Distribución del Puntaje de Felicidad (Score):
El histograma muestra que la mayoría de los países tienen un puntaje de felicidad entre 6 y 7, con una concentración notable en ese rango. El boxplot revela pocos valores atípicos con puntajes extremadamente bajos o altos, lo que indica que la mayoría de los países se agrupan en un rango medio de felicidad, aunque algunos países destacan por ser significativamente más o menos felices.

### Distribución del PIB per Cápita (GDP per capita):
La distribución es asimétrica y sesgada hacia la derecha, lo que indica que un pequeño número de países tiene un PIB per cápita notablemente alto. El boxplot confirma esto, mostrando varios valores atípicos en la cola derecha. La mayoría de los países tienen un PIB per cápita moderado o bajo, con solo unos pocos en niveles significativamente más altos.

### Distribución de la Expectativa de Vida Saludable (Healthy life expectancy):
La distribución es más simétrica que las anteriores, con un pico en torno a 0.9. El boxplot revela algunos valores atípicos en los extremos, lo que indica que la mayoría de los países tienen una expectativa de vida saludable bastante alta, pero unos pocos destacan por tener expectativas más bajas o más altas.

### Distribución de la Libertad para Tomar Decisiones de Vida (Freedom to make life choices):
Al igual que el PIB per cápita, esta variable también muestra una distribución sesgada a la derecha, lo que sugiere que muchos países disfrutan de niveles moderados de libertad, pero unos pocos gozan de niveles mucho más altos. El boxplot muestra varios valores atípicos, destacando algunos países donde la libertad percibida es excepcionalmente alta o baja.

Este análisis refleja cómo las variables se distribuyen y la presencia de valores atípicos en cada caso, proporcionando una comprensión clara de la variabilidad entre países en términos de felicidad, bienestar y factores económicos/sociales.
