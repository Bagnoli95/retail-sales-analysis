```python
# Cargar un DataFrame de ejemplo
df = pd.read_csv('data.csv')

# Identificar valores nulos en todo el DataFrame
print(df.isnull().sum())

# Identificar valores nulos en una columna específica
print(df['columna'].isnull().sum())

# Rellenar valores nulos con interpolación
df['columna'] = df['columna'].interpolate()

# Rellenar valores nulos hacia adelante
df['columna'] = df['columna'].ffill()

# Rellenar valores nulos hacia atrás
df['columna'] = df['columna'].bfill()


from sklearn.linear_model import LinearRegression

# Crear un modelo de regresión para predecir los valores faltantes
model = LinearRegression()

# Seleccionar las filas sin valores nulos para entrenar el modelo
train_data = df.dropna(subset=['columna', 'columna_relacionada'])
X_train = train_data[['columna_relacionada']]
y_train = train_data['columna']

# Entrenar el modelo
model.fit(X_train, y_train)

# Predecir los valores faltantes
missing_data = df[df['columna'].isnull()]
X_missing = missing_data[['columna_relacionada']]
df.loc[df['columna'].isnull(), 'columna'] = model.predict(X_missing)

# Rellenar valores nulos con la media de la columna
df['columna'] = df['columna'].fillna(df['columna'].mean())

# Rellenar valores nulos con la mediana de la columna
df['columna'] = df['columna'].fillna(df['columna'].median())

# Rellenar valores nulos con la moda de la columna
df['columna'] = df['columna'].fillna(df['columna'].mode()[0])

# Eliminar filas con datos faltantes
df = df.dropna()

# Eliminar columnas con datos faltantes
df = df.drop(columns=['columna'])
```
