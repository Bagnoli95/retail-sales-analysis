```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
data = df = pd.read_csv('../data/retail_sales_dataset.csv')

# Convertir la columna 'Date' a tipo datetime
data['Date'] = pd.to_datetime(data['Date'])

# Calcular la correlación
correlation_matrix = data[['Age', 'Quantity', 'Price per Unit', 'Total Amount']].corr()

# Crear un mapa de calor
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa de Calor de Correlación')
plt.show()
```

```python
# Configurar subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Gráfico de Edad por Género
sns.boxplot(x='Gender', y='Age', data=data, ax=axs[0, 0])
axs[0, 0].set_title('Distribución de Edad por Género')
axs[0, 0].grid(True)
axs[0, 0].annotate('Valor alto', xy=(1, 60), xytext=(1, 65), arrowprops=dict(facecolor='black', shrink=0.05), fontsize=12, color='red')

# Gráfico de Cantidad por Género
sns.boxplot(x='Gender', y='Quantity', data=data, ax=axs[0, 1])
axs[0, 1].set_title('Distribución de Cantidad por Género')
axs[0, 1].grid(True)

# Gráfico de Total Amount por Género
sns.boxplot(x='Gender', y='Total Amount', data=data, ax=axs[1, 0])
axs[1, 0].set_title('Distribución de Total Amount por Género')
axs[1, 0].grid(True)

# Gráfico de Precio por Unidad por Género
sns.boxplot(x='Gender', y='Price per Unit', data=data, ax=axs[1, 1])
axs[1, 1].set_title('Distribución de Precio por Unidad por Género')
axs[1, 1].grid(True)

# Ajustar el layout
plt.tight_layout()
plt.show()
```
